```python
from pyModbusTCP.server import DataBank, ModbusServer
import random
from time import sleep
```

```python
# Criando a classe servidor modbus
class ServidorMODBUS():
    """
    Classe servidor Modbus
    """
    
    def __init__(self, host_ip, port):
        """
        Construtor
        """
        
        #Esta tudo encapsulado nessa classe (rotina de atendimento ao cliente)
        self._server = ModbusServer(host=host_ip, port=port, no_block=True)
        self._db = DataBank
    
    def run(self):
        """
        Execução do servidor modbus
        """
        try:
            self._server.start()
            print("Em execução")
            while True:
                self._db.set_words(1000, [random.randrange(int(0.95*400), int(1.05*400))])
                self._db.set_bits(1001,[random.randrange(int(0), int(2))])
                print("==================")
                print("tabela MODEBUS")
                print("Holding register R1000:", self._db.get_words(1000))
                print("Holding register R2000:", self._db.get_words(2000))
                print('Coil R1000:', self._db.get_bits(1001))
                sleep(1)
        except Exception as e:
            print("Erro: ", e.args)
```

```python
s = ServidorMODBUS('localhost', 502)
s.run()
```
